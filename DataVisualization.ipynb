```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
from google.colab import files 
upload=files.upload()
df=pd.read_csv('vaccination-data(26-08-2021).csv')
```

```python
plt.bar(df.WHO_REGION,df.TOTAL_VACCINATIONS,color="red")
plt.ylabel("TOTAL_VACCINATIONS")
plt.xlabel("REGION")
plt.title("VACCINATIONS DISTRIBUTION")
plt.ylabel("TOTAL_VACCINATIONS",fontdict={'fontsize':16,'fontweight':4,'color':"black"})
plt.xlabel("REGION",fontdict={'fontsize':16,'fontweight':4,'color':"black"})
```
